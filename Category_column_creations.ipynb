# Category Column creation

```python
df['sex'] = df['sex']
```

```python
Female = [0]
Male = [1]
```

```python
df['sex'].unique()
```

```python
#this code only works for boolean datatype so we didnt use it. we used the code below this one.
for i in df["sex"]:
  if i in Female:
    df["sex"].replace([i], "Female", inplace=True)
  elif i in Male:
    df["sex"].replace([i], "Male", inplace=True)
```

```python
#the datatype was int64 not boolean
#we overwritten our original column. which is not cool. remember to create a new column instead next time.
df["sex"] = df["sex"].astype(str).replace({ "1": "Male", "0": "Female" })
df
```
